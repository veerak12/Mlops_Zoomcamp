```python
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# data_jan and data_feb are  defined and loading with appropriate columns

# Convert datetime columns to pandas datetime
data_jan['tpep_pickup_datetime'] = pd.to_datetime(data_jan['tpep_pickup_datetime'])
data_jan['tpep_dropoff_datetime'] = pd.to_datetime(data_jan['tpep_dropoff_datetime'])
data_feb['tpep_pickup_datetime'] = pd.to_datetime(data_feb['tpep_pickup_datetime'])
data_feb['tpep_dropoff_datetime'] = pd.to_datetime(data_feb['tpep_dropoff_datetime'])

# Compute duration in minutes
data_jan['duration'] = (data_jan['tpep_dropoff_datetime'] - data_jan['tpep_pickup_datetime']).dt.total_seconds() / 60
data_feb['duration'] = (data_feb['tpep_dropoff_datetime'] - data_feb['tpep_pickup_datetime']).dt.total_seconds() / 60

# Filter data to keep only records with duration between 1 and 60 minutes
data_jan_filtered = data_jan[(data_jan['duration'] >= 1) & (data_jan['duration'] <= 60)].copy()
data_feb_filtered = data_feb[(data_feb['duration'] >= 1) & (data_feb['duration'] <= 60)].copy()

# Convert categorical features to string
categorical = ['PULocationID', 'DOLocationID']
data_jan_filtered[categorical] = data_jan_filtered[categorical].astype(str)
data_feb_filtered[categorical] = data_feb_filtered[categorical].astype(str)

# Combine training and validation data for fitting the vectorizer
combined_data = pd.concat([data_jan_filtered[categorical], data_feb_filtered[categorical]])

# Turn the DataFrame into a list of dictionaries
combined_dicts = combined_data.to_dict(orient='records')

# Fit a dictionary vectorizer on the combined data
dv = DictVectorizer()
dv.fit(combined_dicts)

# Transform training and validation data
feature_matrix_train = dv.transform(data_jan_filtered[categorical].to_dict(orient='records'))
feature_matrix_val = dv.transform(data_feb_filtered[categorical].to_dict(orient='records'))

# Assign the target variables
y_train = data_jan_filtered['duration'].values
y_val = data_feb_filtered['duration'].values

# Train a linear regression model
lr = LinearRegression()
lr.fit(feature_matrix_train, y_train)

# Predict on the training and validation data
y_pred_train = lr.predict(feature_matrix_train)
y_pred_val = lr.predict(feature_matrix_val)

# Calculate RMSE on training and validation data
rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
rmse_val = np.sqrt(mean_squared_error(y_val, y_pred_val))

print("RMSE on training data:", rmse_train)
print("RMSE on validation data:", rmse_val)
```
